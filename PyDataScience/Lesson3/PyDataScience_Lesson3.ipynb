```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

# Task 1. Plot functions

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([3.5, 3.8, 4.2, 4.5, 5, 5.5, 7])

plt.plot(x, y)

plt.show()
```

```python
# Task 2. Plot views

t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.plot(t, f, -)
plt.title('График f(t)', fontsize = 16, fontweight = "bold", loc = "right")

plt.axis([0.5, 9.5, -2.5, 2.5])
```

```python
# Task3. Cards statistics
import pandas as pd

cards = pd.read_csv("C:/Develop/Some/MyProjects/Aux/creditcard.csv ", sep=",")
vals = cards["Class"].value_counts()

vals.plot(kind="bar", logy=True)
```

```python
# Task4. Cards statistics

class0 = cards.loc[cards['Class'] == 0]['V1']
class1 = cards.loc[cards['Class'] == 1]['V1']

plt.hist(class0, bins=20, alpha=0.5, label='Class0', color = 'Gray', density = True)
plt.hist(class1, bins=20, alpha=0.5, label='Class1', color = 'Red', density = True)
plt.legend(loc='upper left')

axis_font = {
"fontsize": 12,
"fontweight": "bold",
"family": "serif",
}

plt.xlabel('V1', axis_font)
plt.show()
```
